```python
path = "/content/drive/MyDrive/Projetos/data"
df_train = pd.read_csv(f'{path}/train.csv')
df_test = pd.read_csv(f'{path}/test.csv')
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
df_train = df_train.drop('ID_code', axis=1)
```

```python
X_train, y_train = df_train.drop('target', axis=1), df_train['target']
```

```python
all(df_train.isnull().sum()) == False
```

```python
pca = PCA(n_components=0.95)
X_reduced = pca.fit_transform(X_train)
```

```python
X_reduced
```

```python
componentes = pca.components_
componentes
```

```python
import joblib
joblib.dump(pca, f'{path}/pca_model.pkl')
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
clf = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0,
    max_depth=1, random_state=0).fit(X_train, y_train)
```

```python
clf.predict(df_test.drop("ID_code", axis=1))
```

```python
joblib.dump(clf, f'{path}/xgb_model.pkl')
```
